Problem Statement

Write code to position 5 equal-sized (50 X 50px) boxes A, B, C, D, and E as followsC
1. box A 200px from the left, and 200px from the top of the viewport.
2. box B -30px left, and -30px above from the center of box A.
3. box B -30px right, and -30px above from the center of box A.
4. box B -30px left, and -30px below from the center of box A.
5. box B -30px right, and -30px below from the center of box A.

```python
<!DOCTYPE html>
<html>
<head>
<style>
  .box {
    width: 50px;
    height: 50px;
    position: absolute;
    background-color: aquamarine;
    text-align: center;
    line-height: 50px;
    font-weight: bold;
    border: 1px solid #000;
  }

  /* Box A at fixed position */
  #A {
    left: 200px;
    top: 200px;
  }

  /* Box B: 30px left, -30px above from center of A */
  #B {
    left: calc(200px - 30px - 50px);
    top: calc(200px + 25px - 30px - 50px / 2);
  }

  /* Box C: 30px right, -30px above from center of A */
  #C {
    left: calc(200px + 50px + 30px);
    top: calc(200px + 25px - 30px - 50px / 2);
  }

  /* Box D: 30px left, -30px below from center of A */
  #D {
    left: calc(200px - 30px - 50px);
    top: calc(200px + 25px + 30px - 50px / 2);
  }

  /* Box E: 30px right, -30px below from center of A */
  #E {
    left: calc(200px + 50px + 30px);
    top: calc(200px + 25px + 30px - 50px / 2);
  }
</style>
</head>
<body>

<div id="A" class="box">A</div>
<div id="B" class="box">B</div>
<div id="C" class="box">C</div>
<div id="D" class="box">D</div>
<div id="E" class="box">E</div>

</body>
</html>
```

Problem Statement

 Explain the difference between Absolute and Relative positioning.
 - **Absolute Positioning**
- **Definition:** An element with position: absolute is positioned relative to the **nearest positioned ancestor** (the closest parent with position set to relative, absolute, or fixed).

  - If no positioned ancestor exists, it’s positioned relative to the **initial containing block** (usually the <html> or <body>).

- **Behavior:**

 -  It is removed from the normal document flow (it doesn’t affect or get affected by the positions of other elements in the normal flow).

 -  You can place it exactly where you want using top, left, right, bottom.

- **Example:**

```python
#css
.absolute-box {
  position: absolute;
  top: 50px;
  left: 100px;
}
```

- **Relative Positioning**
  - **Definition:** An element with position: relative is positioned relative to its normal position in the flow of the document.

- **Behavior:**

  - It stays in the normal flow, and space for it is still preserved.

  - You can nudge it around using top, left, right, bottom, and it will move relative to where it would have been.

  - Often used as a positioning anchor for absolutely positioned child elements.

- **Example:**

```python
#css
.relative-box {
  position: relative;
  top: 10px; /* Moves it down 10px from where it normally would be */
  left: 20px;
}
```

Task 5:


Problem Statement

Explain the z-index, with a code example.
- **z-index:**
  - The z-index CSS property controls the stacking order of elements on a webpage — basically, which element appears in front when elements overlap.

  - Higher z-index → appears on top of elements with lower values.

  - Works only on elements that have a positioning context (position: relative, absolute, fixed, or sticky).

- **Key Points:**
  - Default value is auto (which acts like 0 unless otherwise set).

  - Elements with the same z-index follow the order they appear in the HTML (later ones appear on top).

  - Negative values (z-index: -1) place an element behind others.
- **Code Example:**  





```python
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Z-Index Example</title>
  <link rel="stylesheet" href="c.css">
</head>
<body>

<div class="box red">Red</div>
<div class="box blue">Blue</div>
<div class="box green">Green</div>

</body>
</html>
```

```python
.box {
  width: 150px;
  height: 150px;
  position: absolute;
  top: 50px;
  left: 50px;
  color: white;
  font-size: 20px;
  text-align: center;
  line-height: 150px;
}

/* Middle layer */
.red {
  background-color: red;
  z-index: 1;
}

/* Top layer */
.blue {
  background-color: blue;
  top: 80px;
  left: 80px;
  z-index: 2;
}

/* Bottom layer */
.green {
  background-color: green;
  top: 110px;
  left: 110px;
  z-index: 0;
}
```
